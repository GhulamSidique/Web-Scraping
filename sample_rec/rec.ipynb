```python
import cv2
import numpy as np
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time
import os

def open_url_headless(url, headless = False):
    chrome_options = Options()
    if headless:
        chrome_options.add_argument("--headless")  # Background mode
    chrome_options.add_argument("--window-size=1920x1080")
    chrome_options.add_argument("--disable-gpu")
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")
    chrome_options.add_argument("--hide-scrollbars")
    
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=chrome_options)
    driver.get(url)
    
    try:
        provided_u_name = "haider1805@icloud.com"
        user_name_id = "username"
        user_name = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.NAME, user_name_id)))   
        user_name.send_keys(provided_u_name)
    except Exception as e:
        print("No username tab found")
    # get password tab
    try:
        provided_pass = "Muhssan7865"
        pass_id = "password"
        password = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.NAME, pass_id)))
        password.send_keys(provided_pass)
    except Exception as e:
        print("No password tab found") 
     # get the login tab to enter the username and password
    try:
        # login_tab_css = "submit"
        login_tab = driver.find_element(By.CLASS_NAME,  "onelogin-main__btn")
        driver.execute_script("arguments[0].scrollIntoView();", login_tab)
        login_tab.click()
    except Exception as e:
        print("No login tab found here")
    
    # =============================================================================================
    # Handle cookie consent banner
    try:
        # cookie_main_bar = driver.find_element(By.ID, "onetrust-button-group")
        cookie_accept = driver.find_element(By.ID, "onetrust-reject-all-handler")
        driver.execute_script("arguments[0].scrollIntoView(true);", cookie_accept)
        time.sleep(1)
        cookie_accept.click()
    except Exception as e:
        print(f"Error handling cookie banner:")

    time.sleep(5)
    return driver

def record_website_screen(driver, output_file="website_record.avi", fps=20):
    img = driver.get_screenshot_as_png()
    nparr = np.frombuffer(img, np.uint8)
    frame = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
    height, width, _ = frame.shape
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(output_file, fourcc, fps, (width, height))

    # start_time = time.time()
    try:
        while True: #time.time() - start_time < duration:
            img = driver.get_screenshot_as_png()
            nparr = np.frombuffer(img, np.uint8)
            frame = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
            out.write(frame)
            time.sleep(1/fps)
    except KeyboardInterrupt:
        print("Recording interrupted.")
    finally:
        out.release()
        print("Recording saved!")

if __name__ == "__main__":
    url = "https://www.bca.co.uk/sales/live?catalogueRef=CO073Z"
    driver = open_url_headless(url, headless=False)  # Set headless=False to see the tab
    print("Recording in background...")
    record_website_screen(driver, output_file="website_background_record.avi", fps=10)
    driver.quit()
    print("Recording saved!")
```
