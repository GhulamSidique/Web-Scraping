### download the images

```python
from urllib.parse import urlparse, urljoin

# urlparse: Parses a URL into components (scheme, netloc, path, etc.), making it easy to validate or extract information from the URL.
# urljoin: Joins a base URL with a relative path to form a complete URL.

def download_images(data, main_folder="Images"):  # The main folder is "Images"
    
    # Create the main folder if it doesn't exist
    os.makedirs(main_folder, exist_ok=True)
    
    # Loop over every row to get the info
    for index, row in data.iterrows():
        reg_no = row["Title"]  # Separate reg nums
        image_urls = row["Images"].split("; ")  # Split image URLs by comma if multiple
        
        # Create a subfolder for the car registration number
        reg_folder = os.path.join(main_folder, reg_no)  # Combine main folder name and reg num
        os.makedirs(reg_folder, exist_ok=True)  # Create the subfolder if it doesn't exist
        
        # Loop through the URLs of the current row
        for idx, url in enumerate(image_urls):
            url = url.strip()  # Remove extra spaces
            
            if not url.startswith(("http://", "https://")):  # Check if the URL does not start with "http" or "https"
                url = urljoin("https://", url)  # Ensure URLs start with "https://"
            
            # Parse the URL
            parsed_url = urlparse(url)

            # Check if the parsed URL is valid
            if not parsed_url.scheme or not parsed_url.netloc:
                print(f"Invalid URL skipped: {url}")  # Skip invalid URLs
                continue
            
            # Try downloading images
            try:
                # Download the image
                response = requests.get(url, stream=True)  # Send request to download image
                response.raise_for_status()  # Raise an exception if the HTTP request fails
                
                # Extract file name from URL path
                file_name = os.path.basename(parsed_url.path)

                # Remove any query parameters or extra text after the valid extension
                base_name, file_extension = os.path.splitext(file_name)
                valid_extensions = [".jpg", ".jpeg", ".png", ".gif", ".bmp", ".webp"]
                
                # Ensure it has a valid image extension
                if file_extension.lower() not in valid_extensions:
                    file_extension = ".jpg"  # Default to .jpg if extension is not valid
                
                file_name = base_name.split("?")[0] + file_extension  # Remove anything after "?" in the base name
                
                # Construct the full path for the file
                full_file_name = os.path.join(reg_folder, f"{reg_no}_{idx + 1}_{file_name}")
                
                # Save the image
                with open(full_file_name, 'wb') as f:
                    for chunk in response.iter_content(1024):  # Reads the response in 1KB chunks to save memory.
                        f.write(chunk)
                
                print(f"Downloaded: {full_file_name}")
            
            except Exception as e:
                print(f"Failed to download {url} for {reg_no}: {e}")

# Call the function
download_images(df[['Title', 'Images']])
```
